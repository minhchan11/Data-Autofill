```python
# UNBALANCED VERSION

import statistics 
import math
dictionary = {'product quality':4,'high shipping fee':5,'invalid complaint':10,'long delivery leadtime':16,\
               'return policy':4,'received items issue':9,'seller service':8, 'check upon delivery':20,\
              'false negative':80, 'shipping policy':27, 'product information':72, 'seller attitude': 58,\
              'cancellation':70, 'delivery service':36, 'carrier attitude':25, 'campaigns':52, 'app issues':38}
 #categorize() #apply the categorize method => input: {cat:added value}
input = {'high shipping fee':23,'invalid complaint':30,'long delivery leadtime':41,\
               'return policy':30,'received items issue':25,'seller service':10, 'check upon delivery':27,\
              'false negative':53, 'shipping policy':39, 'product information':28, 'seller attitude': 34,\
              'cancellation':43, 'delivery service':35, 'carrier attitude':28, 'campaigns':32, 'app issues':48}

 #while  :
sorteddict = sorted(dictionary, key=dictionary.get,reverse=True)
for cat in input.keys():
    if cat not in sorteddict:
        sorteddict.append(cat)
        dictionary[cat]=0
total = sum(dictionary.values())
if dictionary[sorteddict[3]]>total*0.1:
    medval = math.ceil(statistics.mean(dictionary.values()))
else:
    list = []
    for cat in sorteddict[:5]:
        list.append(dictionary[cat])
        medval = math.ceil(statistics.median(list))
for cat in sorteddict[:]:
    if cat in input.keys():
        dictionary[cat]+=input[cat]
    print(cat+': '+str(dictionary[cat]))
```

```python
# BALANCED VERSION

import statistics
import math
dictionary = {'product quality':4,'high shipping fee':5,'invalid complaint':10,'long delivery leadtime':16,\
               'return policy':4,'received items issue':9,'seller service':8, 'check upon delivery':20,\
              'false negative':80, 'shipping policy':27, 'product information':72, 'seller attitude': 58,\
              'cancellation':70, 'delivery service':36, 'carrier attitude':25, 'campaigns':52, 'app issues':38}
 #categorize() #apply the categorize method => input: {cat:added value}
input = {'high shipping fee':23,'invalid complaint':30,'long delivery leadtime':41,\
               'return policy':30,'received items issue':25,'seller service':10, 'check upon delivery':27,\
              'false negative':53, 'shipping policy':39, 'product information':28, 'seller attitude': 34,\
              'cancellation':43, 'delivery service':35, 'carrier attitude':28, 'campaigns':32, 'app issues':40}

 #while  :
sorteddict = sorted(dictionary, key=dictionary.get,reverse=True)
for cat in input.keys():
    if cat not in sorteddict:
        sorteddict.append(cat)
        dictionary[cat]=0
total = sum(dictionary.values())
if dictionary[sorteddict[4]]>total*0.1:
    medval = math.ceil(statistics.mean(dictionary.values()))
else:
    list = []
    for cat in sorteddict[:5]:
        list.append(dictionary[cat])
        medval = math.ceil(statistics.median(list))
print ('first five:')
for cat in sorteddict[:5]:
    print(cat+': '+str(dictionary[cat]))
print ('\n\nnext 7:')    
for cat in sorteddict[5:12]:
    if cat in input.keys():
        if dictionary[cat]+input[cat]<medval:
            dictionary[cat]+=input[cat]
        else:
            dictionary[cat]=medval
    print(cat+': '+str(dictionary[cat]))
print ('\n\nremaining:')
for cat in sorteddict[12:]:
    if cat in input.keys():
        dictionary[cat]+=input[cat]
    print(cat+': '+str(dictionary[cat]))
        
```
